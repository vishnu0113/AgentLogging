Observations:

```python
df1.info()
```

Obersvations:
1. df1 has 3 int, 1 float and 3 object Datatype
2. Average Response Time & Resolution Time is to be converted to int data type

```python
df['Date']
```

```python
#segregating date, month and year from the Date dataframe
df['Log_date']= df['Date'].str.split('-').str[0]
df['Log_month']= df['Date'].str.split('-').str[1]
df['Log_year']= df['Date'].str.split('-').str[2]
```

```python
df
```

```python
#converting logging date and year in int datatype
#as month is string is not converted to int 
df['Log_date']= df['Log_date'].astype(int)
df['Log_year']= df['Log_year'].astype(int)
```

```python
#removing the Date dataframe
df.drop('Date', axis=1, inplace=True)
df.head(10)
```

```python
df.dtypes
```

```python
df['Duration']
```

```python
#segregation of Duration into Duration hour, min secs
df['Duration_hr']= df['Duration'].str.split(':').str[0]
df['Duration_min']= df['Duration'].str.split(':').str[1]
df['Duration_sec']= df['Duration'].str.split(':').str[2]
```

```python
#converting the data in int 
df['Duration_hr']= df['Duration_hr'].astype(int)
df['Duration_min']= df['Duration_min'].astype(int)
df['Duration_sec']= df['Duration_sec'].astype(int)
```

```python
#removing Duration column
df.drop('Duration', axis=1, inplace=True)
```
